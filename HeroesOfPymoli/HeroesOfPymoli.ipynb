```python
#find number of purchases & total dollars in purchases by gender
purchase_count_by_gen = purchase_data.groupby(by="Gender").nunique()
purchase_tot_by_gen = purchase_data.groupby(by="Gender").sum()

genders_alpha = ["Female", "Male", "Other / Nondisclosed"]
purchase_count = []
avg_purch_price = []
total_purch_val = []
avg_tot_purch = []

#put data into lists
for i in range(3):
    purchase_count.append(purchase_count_by_gen.PurchaseID[i])

for i in range(3):
    total_purch_val.append(purchase_tot_by_gen.Price[i])
    
total_purch_rounded = ["$" + str(round(i, 2)) for i in total_purch_val]

#nested loop to find average purchase amount
for i in purchase_count:
    for j in total_purch_val:
        if purchase_count.index(i) == total_purch_val.index(j):
            avg_purch_price.append(j/i)
        continue
        
avg_purch_rounded = ["$" + str(round(i, 2)) for i in avg_purch_price]

#nested loop to find average total purchased per person for each gender
for i in gender_counts:
    for j in total_purch_val:
        if gender_counts.index(i) == total_purch_val.index(j):
            avg_tot_purch.append(j/i)
        continue
        
avg_tot_purch_rounded = ["$" + str(round(i, 2)) for i in avg_tot_purch]

gender_purch_analysis_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": avg_purch_rounded, "Total Purchase Value": total_purch_rounded, "Avg Total Purchase per Person": avg_tot_purch_rounded}, genders_alpha)
gender_purch_analysis_df

```

```python
#group dataframe by ScreenName to find out how much each player spends
big_spenders = purchase_data.groupby("SN").agg({"Price": "sum", "PurchaseID": "count"})
#then sort and find top five
top_5 = big_spenders.sort_values(by="Price", ascending=False).head()

top_5_tot = []
top_5_SN = []

#put the data from the 'Groupby'Dataframe into lists
for i in top_5.index:
    top_5_SN.append(i)
top_5_SN

for i in range(5):
    top_5_tot.append(top_5.Price[i])
    
num_purchases = []    

for i in range(5):
    num_purchases.append(top_5.PurchaseID[i])
    
top_5_avg_purch = []

#find the average purchase made by a player in the top 5
for i in top_5_tot:
    for j in num_purchases:
        if num_purchases.index(j) == top_5_tot.index(i):
            top_5_avg_purch.append(i/j)
        continue
top_5_avg_round = ["$" + str(round(i,2)) for i in top_5_avg_purch]
top_5_tot_round = ["$" + str(round(i,2)) for i in top_5_tot]

big_spenders_df = pd.DataFrame({"Purchase Count": num_purchases, "Average Purchase Price": top_5_avg_round, "Total Purchase Valie": top_5_tot_round}, top_5_SN)
big_spenders_df
```
